## Integer

- Used to store whole numbers

- Can be positive, negative, or zero

- No decimal point

```python
x=10
y=-20
z=0
```

```python
print(type(x),type(y),type(z))
```

## Float

- Used to store decimal numbers

- Can be positive or negative

- Always contains a decimal point

```python
a=10.2
b=-5.1234
c=0.0
```

```python
print(type(a), type(b), type(c))
```

## String

- Used to store text

- Written inside single ' ' or double " " quotes


```python
name = "Roger"
message = 'Hello'
```

```python
print(type(name), type(message))
```

## complex

- used to store complex numbers in the form `a + bj`
- a â†’ Real part
- b â†’ Imaginary part

```python
x = 2 + 3j
```

```python
x.real
```

```python
x.imag
```

```python
y=3 - 7j
```

```python
y.real
```

```python
y.imag
```

### Note: 

- only `j` is recognised as imaginary part
- `j` should be written after the number not before

```python
z=-3+j9
```

```python
z=-3+9i
```

```python
z=-3+9j
```

```python
z.real
```

```python
z.imag
```
