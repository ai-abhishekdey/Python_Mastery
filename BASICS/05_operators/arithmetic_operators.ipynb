## Arithmetic Operators

Used for mathematical calculations.

### Addition (+)

```python
a=20
b=10

c=a+b

print(c)
```

### Substraction (-)

```python
a=20
b=10

c=a-b

print(c)
```

### Multiplication (*)

```python
a=20
b=10

c=a*b

print(c)
```

### Exponent or Power (**)

```python

a=10**3
b=5**2

print(f"Value of a is {a}")
print(f"Value of b is {b}")
```

### Division (/)

- Always returns float.

```python
a = 100 / 2

b = 100 / 3

print(f"Value of a is {a}")
print(f"Value of b is {b}")
```

### Floor Division (//)

- Removes decimal part.

```python
a = 100 // 2

b = 100 // 3

print(f"Value of a is {a}")
print(f"Value of b is {b}")
```

### Modulus (%)

- Returns remainder

```python
a = 100 % 2

b = 100 % 3

print(f"Value of a is {a}")
print(f"Value of b is {b}")
```
