## Assignment Operators

### Note:

```
x += y  -> x = x + y
x -= y  -> x = x - y
x *= y  -> x = x * y
x /= y  -> x = x / y 
x //= y -> x = x // y
x **= y -> x = x ** y
```

### Add and Assign (+=)

````python
x = 10
x += 3

print(x)
````

### Substract and assign

````python
a = 20
a -= 5

print(a)
````

### Multiply and assign

````python
b = 4
b *= 5

print(b)
````

### Divide and assign

````python
c = 45
c /=10

print(c)
````

### Floor Divide and assign

````python
d = 45
d //= 10

print(d)
````

### Modulus and assign

````python
p = 25
p %= 6

print(p)
````

### Power and assign

````python
K = 4
K **= 2

print(K)
````
