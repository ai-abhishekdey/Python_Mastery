## Comparison Operators

Comparison operators are used to compare two values.

They always return: `True` or `False`

### Example

```python
a = 10
b = 20
c = 10
```

### Equal to (==)

```python
a == b
```

```python
a == c
```

### Not Equal to (!=)

```python
a != c
```

```python
a != b
```

### Less than (<)

```python
a < b
```

```python
a < c
```

### Less than equal to (<=)

```python
a <= b
```

```python
a <= c
```

### Greater than (>)

```python
a > b
```

```python
a > c
```

```python
b > a
```

### Greater than equal to (>=)

```python
a >= b
```

```python
a >= c
```

```python
b >= a
```
