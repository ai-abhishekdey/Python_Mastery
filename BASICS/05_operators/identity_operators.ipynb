## Identity Operators

- Identity operators are used to check whether two variables refer to the same object in memory.

- There are two identity operators:

    - is

    - is not

### How to Check Memory Location of a Variable in Python

- In Python, you can check the memory address of an object using: `id() function`

## Examples

```python
x = 20
y = 20
```

```python
id(x) # memory location of x
```

```python
id(y) # memory location of y
```

```python
print(x is y) # True since memory location x and y are same
```

```python
a = [1,2,3]
b = [1,2,3]
```

```python
print(a is b)
```

### Why False ? Lets see the memory locations

```python
id(a)
```

```python
id(b)
```

#### Since a and b are pointing to different memory locations, a is b -> False


```python
print(a is not b)
```

#### Similarly Since a and b are pointing to different memory locations, a is not b -> True
