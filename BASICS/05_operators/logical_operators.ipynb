## Logical Operators

Logical operators are used to combine multiple conditions.

There are three logical operators in Python:

- and

- or

- not

### `and` Operator

- Returns True only if both conditions are True.

```python
a=20

b = a > 5 and a < 30 # both are true so output is true
c = a < 40 and a > 25 # only one is true so output is false
d = a > 25 and a > 45 # both are false so output is flase

print(f" b = {b}")
print(f" c = {c}")
print(f" d = {d}")
```

### `or` Operator

- Returns True if at least one condition is True

```python
a=20

b = a > 5 or a < 30 # both are true so output is true
c = a < 40 or a > 25 # only one is true so output is true
d = a > 25 or a > 45 # both are false so output is flase

print(f" b = {b}")
print(f" c = {c}")
print(f" d = {d}")
```

### `not` Operator

- reverses the result

```python
a=20

b = not (a > 5 or a < 30) # both are true so output is true -> reverse = false
c = not (a < 40 or a > 25) # only one is true so output is true -> reverse = false
d = not (a > 25 or a > 45) # both are false so output is false -> reverse = True

print(f" b = {b}")
print(f" c = {c}")
print(f" d = {d}")
```
