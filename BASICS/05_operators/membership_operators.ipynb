## Membership Operators

Membership operators are used to check whether a value exists inside a sequence.

- There are two membership operators:

    - in

    - not in

- They always return True or False.

### With lists

```python
a=[1,2,3,4,5]

print(3 in a)
```

```python
print(10 in a)
```

```python
print(2 not in a)
```

```python
print(20 not in a)
```

```python
b = ["castle", "building", "summation"]
```

```python
print("build" in b)
```

```python
print("building" in b)
```

```python
print("cas" not in b)
```

```python
print("castle" not in b)
```

### With Range

```python
k = range(10)
```

```python
print(5 in k)
```

```python
print(20 in k)
```

```python
print(4 not in k)
```

```python
print(47 not in k)
```

### With Tuple

```python
print(20 not in t)
```

```python
k = ("cycling", "batting", "crying", 30, 2)
```

```python
print("batting" in k)
```

```python
print("bat" in k)
```
