## Python loops

### Range Funtion

- Syntax: range(start, stop-1, step)

- If start is not specified, start=0 is considered

- If step is not specified, step=1 is considered


### For loop with range function - (start, stop, step)

````python
for i in range(0,5,1):
    print(i)
````

### For loop with range function - (start, stop)

````python
for i in range(0,5):
    print(i)
````

### For loop with range function - (only stop)

````python
for i in range(5):
    print(i)
````

### Negative step - Reverse loop

````python
for i in range(10,2,-2):
    print(i)
````

### For loop with Lists

````python
fruits=["Apple", "Banana", "Mango", "Guava"]

for fruit in fruits:
    print(fruit)
````

````python
scores=[0.4, 0.67, 0.90, 0.39]

for score in scores:
    print(score)
````

### For loop with list and range

````python
fruits=["Apple", "Banana", "Mango", "Guava"]

for i in range(len(fruits)):
    
    fruit=fruits[i]

    print(f"index : {i} -> fruit : {fruit}")
````

### For loop with Tuple

````python
cars=("Maruti", "Tata",  "hyundai")

for car in cars:
    
    print(car)
````

````python
numbers = (23, 98.5, 103.3, 89)

for number in numbers:
    
    print(number)
````

### For loop with tuple and range

````python
vegetables = ("potato", "tomato", "brinjal", "cauliflower")

for i in range(len(vegetables)):

    vegetable=vegetables[i]

    print(f"index : {i} -> vegetable : {vegetable}")
````

# Nested For Loop

```
For every one iteration of the outer loop,
    the inner loop runs completely.
```    

````python
a = [1,2,3]

b = [10, 20]

for i in a:
    for j in b:
        print(f"i = {i}, j = {j}")
````

````python
for i in range(5):
    for j in range(2):
        print(f"i = {i}, j = {j}")
````

````python
fruits=["Apple", "Banana", "Mango", "Guava"]
vegetables =("potato", "tomato", "brinjal", "cauliflower")

for fruit in fruits:
    for vegetable in vegetables:

        print(f"fruit : {fruit}, veg : {vegetable}")
````
